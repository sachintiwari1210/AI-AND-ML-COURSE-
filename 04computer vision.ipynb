```python
r = img_plt[:,:,0]
g = img_plt[:,:,1]
b = img_plt[:,:,2]
```

```python
r.shape, g.shape, b.shape
```

```python
plt.subplot(1,3,1)
plt.imshow(r, cmap='Reds')
plt.subplot(1,3,2)
plt.imshow(g, cmap='Greens')
plt.subplot(1,3,3)
plt.imshow(b, cmap='Blues')
```

```python
new_img = (r + g + b) / 3
```

```python
plt.imshow(new_img, cmap='gray')
```
