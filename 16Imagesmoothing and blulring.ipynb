##### Avereging Kernel

```python
import cv2 
import numpy as np 

# Load an image 
image = cv2.imread("images/lenna.png")

# Define a 3x3 averaging kernel
kernel = np.ones((5, 5), np.float32) / 25

# Apply the kernel to smooth the image
smoothed_image  = cv2.filter2D(image, -1, kernel)

#Show the original and smoohted images
cv2.imshow("Original", image)
cv2.imshow("Smoothed", smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Gausian Blue Kernel

```python
import matplotlib.pyplot as plt
gray = cv2.imread("images/lenna.png", 0)
kernel_size = 5  #Ajust as needed
kernel = cv2.getGaussianKernel(kernel_size, 1.25)
blurred = cv2.filter2D(gray, -1, kernel) # Apply filter
```

```python
kernel
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('gray image')
plt.imshow(gray)

plt.subplot(1,2,2)
plt.title('Gaussian Blurred Image')
plt.imshow(blurred)
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(121), plt.imshow(gray, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(blurred, cmap='gray'), plt.title('Gaussian Blur')
plt.show()
```
