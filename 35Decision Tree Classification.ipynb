##### IMPORT THE LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##### LOAD THE DATA

```python
df = pd.read_csv('dataset/banknotes.csv')
```

```python
df
```

##### SEPERATE THE INPUT AND OUTPUT VARIABLE

```python
# input features 
x = df.drop('Class', axis = 1)

# output labels
y = df['Class']
```

```python
x.describe()
```

```python
##### EXPLORE THE DATA
```

```python
y.value_counts()
```

```python
plt.pie(y.value_counts(), labels=y.value_counts().index, autopct='%2.2f%%');
```

```python
sns.jointplot(x='Variance',y='Skewness',data=df, hue='Class')
```

```python
sns.pairplot(df, hue='Class')
```

##### CROSS VALIDATION

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

##### BUILT A MODEL

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf_tree = DecisionTreeClassifier()
```

```python
# Train the algorithm
clf_tree.fit(x_train, y_train)
```

##### PERFORMANCE EVALUTION

##### Predic tht eunseen data

```python
y_pred = clf_tree.predict(x_test)
```

```python
y_pred
```

```python
result = pd.DataFrame({
    'actual': y_test,
    'predicted': y_pred
})
result
```

##### CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

##### VISUALIZE THE TREE

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf_tree)
```

```python
plt.figure(figsize=(16,12))
plot_tree(clf_tree, fontsize=6, feature_names=x.columns, filled = True, precision=2, class_names=['0','F'], rounded= True);
```

```python
plt.title('Feature Importances')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.bar(x.columns, clf_tree.feature_importances_, color='navy')
```
