```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('images/lenna.png', 0)
```

```python
new_img = img > 127
```

```python
new_img
```

```python
new_img = img > 127
inv_img = img < 127
```

```python
plt.figure(figsize=(16,9))
plt.subplot(131)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(132)
plt.title('Binary Image')
plt.imshow(new_img, cmap='gray')
plt.subplot(133)
plt.title('Inv Binary Image')
plt.imshow(inv_img, cmap='gray')
```

##### Otsue Threshold

```python
# Apply Otsu's thrsholding
thresh_val, thresh_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)
```

```python
thresh_val
```

```python
plt.figure(figsize=(16,9))
plt.subplot(141)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(142)
plt.title('Binary Image')
plt.imshow(new_img, cmap = 'gray')
plt.subplot(143)
plt.title('Inverse Binary Image')
plt.imshow(inv_img, cmap = 'gray')
plt.subplot(144)
plt.title('Otsu Threshold Image')
plt.imshow(thresh_otsu, cmap = 'gray')
```
