```python
import cv2

image1 = cv2.imread("images/lenna.png")
image2 = cv2.imread("images/sun.png")
```

```python
image1.shape, image2.shape
```

```python
added_image = cv2.add(image1, image2)
cv2.imshow("Added Image", added_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt 
img1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
added = cv2.cvtColor(added_image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(added)
```

##### BLENDED IMAGE

```python
alpha = 0.50 # Transparency factor
blended_image = cv2.addWeighted(img1, alpha, img2, 1-alpha, 0)
```

```python
blended = cv2.cvtColor(blended_image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(img1)

plt.subplot(1,3,2)
plt.imshow(img2)

plt.subplot(1,3,3)
plt.imshow(blended_image)
```

##### Multiplication of images

```python
brightened_image = cv2.multiply(img1, 1.5) # 1.5 TIMES INCREASE BRIGHTNESS
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('original image')
plt.imshow(img1)

plt.subplot(1,2,2)
plt.title('brightened image')
plt.imshow(brightened_image)
```

##### Bitwise AND

```python
img1 = cv2.imread('images/input1.png')
img2 = cv2.imread('images/input2.png')
dest_and = cv2.bitwise_and(img2, img1, mask = None)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(dest_and)
```

##### BITWISE OR

```python
img1 = cv2.imread('images/input1.png')
img2 = cv2.imread('images/input2.png')
dest_or = cv2.bitwise_or(img2, img1, mask = None)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(dest_or)
```

##### BITWISE XOR

```python
img1 = cv2.imread('images/input1.png')
img2 = cv2.imread('images/input2.png')
dest_xor = cv2.bitwise_xor(img2, img1, mask = None)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(dest_xor)
```

##### BITWISE NOT

```python
img1 = cv2.imread('images/input1.png')
img2 = cv2.imread('images/input2.png')
dest_not2 = cv2.bitwise_not(img2)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.imshow(img2)
plt.subplot(1,2,2)
plt.imshow(dest_not2)
```
